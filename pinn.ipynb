```
import plotly.graph_objects as go
import numpy as np
from plotly.subplots import make_subplots
import pandas as pd

fig = make_subplots(rows=2, cols=1, horizontal_spacing=0.051)

df = pd.read_csv(r"/content/drive/My Drive/Colab Notebooks/paper_all_v3.csv")
xs = df[df["t"] == 0]['x']
t = df[df["x"] == -1]['t']
x = -1.0

fig.add_trace(go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_actual'], name="u_actual", \
                         line = dict(color='blue')), row=1, col=1)
fig.add_trace(go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_normal'], name="u_normal", \
                         fill='tonexty', mode="lines", line = dict(color='red')), row=1, col=1)
fig.add_trace(go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_actual'], name="u_actual", \
                         showlegend=False, line = dict(color='red')), row=1, col=1)
fig.add_trace(go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_physics'], name="u_physics", \
                         fill='tonexty', mode="lines", line = dict(color='green')), row=1, col=1)
fig.add_trace(go.Scatter(visible=True, x=t, y=df[df["x"] == x]['du_normal'], name="du_normal", \
                         line = dict(color='red', dash='dash')), row=2, col=1)
fig.add_trace(go.Scatter(visible=True, x=t, y=df[df["x"] == x]['dv_normal'], name="dv_normal", \
                         fill='tonexty', mode="lines", line = dict(color='red', dash='dot')), row=2, col=1)
fig.add_trace(go.Scatter(visible=True, x=t, y=df[df["x"] == x]['du_physics'], name="du_physics", \
                         line = dict(color='green', dash='dash')), row=2, col=1)
fig.add_trace(go.Scatter(visible=True, x=t, y=df[df["x"] == x]['dv_physics'], name="dv_physics", \
                         fill='tonexty', mode="lines", line = dict(color='green', dash='dot')), row=2, col=1)
 

# frames =[go.Frame(data=[go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_actual']),
#                         go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_normal']),
#                         go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_actual']),
#                         go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_physics']),
#                         go.Scatter(visible=True, x=t, y=df[df["x"] == x]['du_normal']),
#                         go.Scatter(visible=True, x=t, y=df[df["x"] == x]['dv_normal']),
#                         go.Scatter(visible=True, x=t, y=df[df["x"] == x]['du_physics']),
#                         go.Scatter(visible=True, x=t, y=df[df["x"] == x]['dv_physics'])],
#                         traces=[0, 1, 2, 3, 4, 5, 6, 7]) for x in xs]

frames = []
for x in xs:  
    frames.append(go.Frame(data=[go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_actual']),
                            go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_normal']),
                            go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_actual']),
                            go.Scatter(visible=True, x=t, y=df[df["x"] == x]['u_physics']),
                            go.Scatter(visible=True, x=t, y=df[df["x"] == x]['du_normal']),
                            go.Scatter(visible=True, x=t, y=df[df["x"] == x]['dv_normal']),
                            go.Scatter(visible=True, x=t, y=df[df["x"] == x]['du_physics']),
                            go.Scatter(visible=True, x=t, y=df[df["x"] == x]['dv_physics'])],
                            name = x,
                            traces=[0, 1, 2, 3, 4, 5, 6, 7]))


fig.frames=frames

play = dict(
             label='Play',
             method='animate',
             args=[None, dict(frame=dict(duration=100, redraw=False), 
                              transition=dict(duration=0),
                              fromcurrent=True,
                              mode='immediate')])
pause = {
                "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                  "mode": "immediate",
                                  "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate"
            }

fig.update_layout(updatemenus=[dict(type='buttons',
                              showactive=False,
                              y=0,
                              x=1.05,
                              xanchor='left',
                              yanchor='bottom',
                              buttons=[play, pause] )
                                      ],
                 width=900, height=600)

steps = []
i = 0             
for x in xs:
    # fig_dict["frames"].append(go.Frame(f[i]))
    
    step = {"args": [
        [x],
        {"frame": {"duration": 300, "redraw": False},
         "mode": "immediate",
         "transition": {"duration": 300}}
    ],
        "label": x,
        "method": "animate"}
    
    steps.append(step)
    i += 1


sliders = {
    "active": 0,
    "yanchor": "top",
    "xanchor": "left",
    "currentvalue": {
        "font": {"size": 20},
        "prefix": "x:",
        "visible": True,
        "xanchor": "right"
    },
    "transition": {"duration": 300, "easing": "cubic-in-out"},
    "pad": {"b": 10, "t": 50},
    "len": 1.0,
    "x": 0,
    "y": 0,
    "steps": steps
}
                              
fig.update_yaxes(range=[-3.5, 3.5], row=1, col=1)
fig.update_yaxes(range=[-3.5, 3.5], row=2, col=1)
fig.update_layout(sliders=[sliders])
fig.show()
```
